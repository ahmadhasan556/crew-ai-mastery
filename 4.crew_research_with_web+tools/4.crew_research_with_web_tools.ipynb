```python
# creating agents
from crewai import Agent , Task , Crew 
from crewai_tools import SerperDevTool
research_agent = Agent(
  role = "Research Specialist",
  goal = "Research interesting facts about {topic} and summarize them in a concise manner.",
  backstory = "You are an expert researcher with a knack for finding and summarizing interesting information.",
  tools = [SerperDevTool()],
  verbose=True,
  llm=llm
)
writer_agent = Agent(
  role = "Creative Writer",
  goal = "Write a short blog summary using the research",
  backstory = "You are skilled at writing engaging and informative blog summaries.",
  verbose=True,
  llm=llm
)
```

```python
#  creating tasks for agent
research_task = Task(
  description="Find 3-5 interesting recent facts about {topic} as of year 2026..",
  expected_output="a bullet list of 3-5 facts",
  agent=research_agent
)
writing_task = Task(
  description="Write a 100-word blog post summary about {topic} using the facts from the research task",
  expected_output="A blog post summary",
  agent=writer_agent,
  context=[research_task]
)
```
